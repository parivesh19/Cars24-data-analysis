# Raw data cleaning using Pandas

```python
# Import libraries
import pandas as pd
```

```python
# Load the data from csv file to df
df=pd.read_csv('cars24_raw_data.csv')
```

```python
df
```

```python
# Drop the column Unnamed
df=df.drop("Unnamed: 0",axis=1)
```

```python
df
```

```python
# Replace comma, space and km and convert it into integer datatype
df['KM_DRIVEN']=df['KM_DRIVEN'].str.replace(',','').str.replace(' km','').astype(int)
```

```python
df
```

```python
# Replace owner and space
df['OWNER_TYPE']=df['OWNER_TYPE'].str.replace(' owner','')
df
```

```python
# Split the data into two parts and extract the first one
df['REGISTRAION_ID']=df['REGISTRAION_ID'].str.split('-').str.get(0)
df
```

```python
# Replace ₹, comma, /month and convert the datatype into float
df['MONTHLY_EMI']=df['MONTHLY_EMI'].str.replace('₹','').str.replace(',','').str.replace('/month','').astype(float)
df
```

```python
# Replace ₹, Lakh and space and convert the datatype into float
df['CAR_PRICE']=df['CAR_PRICE'].str.replace('₹','').str.replace(' Lakh','').astype(float)
df
```

```python
# Replace On and space
df['DOWNPAYMENT_AMOUNT']=df['DOWNPAYMENT_AMOUNT'].str.replace('On ','')
df
```

```python
# Converting the column into string column
df['CAR_MODEL']=df['CAR_MODEL'].astype(str)
```
